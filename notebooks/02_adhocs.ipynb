```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

db_name = 'cg_itemknn.db'
db_path = f'/home/astrofimuk/workspace/mla-contest-recsys-04-2026/artifacts/optuna/{db_name}'

with sqlite3.connect(db_path) as conn:
    query = """
        SELECT p.param_value, v.value
        FROM trial_params p
        JOIN trial_values v ON p.trial_id = v.trial_id
    """
    df = pd.read_sql_query(query, conn)

df.plot(
    x='param_value',
    y='value',
    kind='scatter',
    # figsize=(10, 6),
    title='Зависимость Value от Param Value',
    grid=True
)

plt.show()
```
